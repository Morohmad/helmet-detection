```python
import numpy as np
```

```python
from ultralytics import YOLO

model = YOLO("/content/drive/MyDrive/helmet_detection_project/project/runs/detect/train2/weights/best.pt")
metrics = model.val(data="/content/drive/MyDrive/helmet_detection_project/project/data.yaml")

# Menampilkan metrik rata-rata untuk semua kelas
print("=== EVALUASI KESELURUHAN ===")
print(f"Mean Precision     : {metrics.box.mp:.3f}")
print(f"Mean Recall        : {metrics.box.mr:.3f}")
print(f"mAP@0.5            : {metrics.box.map50:.3f}")
print(f"mAP@0.5:0.95       : {metrics.box.map:.3f}")

# Menampilkan metrik per kelas
print("\n=== EVALUASI PER KELAS ===")
for i, name in enumerate(model.names):
    p, r, ap50, ap = metrics.box.class_result(i)
    print(f"\nClass: {name}")
    print(f"Precision          : {p:.3f}")
    print(f"Recall             : {r:.3f}")
    print(f"mAP@0.5            : {ap50:.3f}")
    print(f"mAP@0.5:0.95       : {ap:.3f}")
```
